# Дана функция f(x) = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
x = Symbol('x', real = True)
y = simplify((sin(x)/2 + cos(x)/2)**2 * sin(x))
y
```

```python
solveset(y, x)
```

```python
solve(y, x)
```

```python
periodicity(y, x)
```

Функция возрастает

```python
derivate = diff(y)
derivate
```

```python
graf = plot(derivate, (x, -pi, pi))
```

```python
solveset(derivate>=0, x)
```

```python
solve(derivate, x, Interval (-pi, pi))
```

Функция убывает

```python
solveset(derivate < 0, x)
```

Построить график

```python
graf = plot(y, (x, -pi, pi))
```

Вычислить вершину


```python
max = maximum(y, x, Interval (-pi, pi)).evalf()
min = minimum(y, x, Interval (-pi, pi)).evalf()

print('max =', round(max, 3))
print('min =', round(min, 3))
```
